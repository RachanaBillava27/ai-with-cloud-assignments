# Assignment-5
##  Student Data Manager

```python
students = {}

# Input for 5 students
for i in range(5):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks

topper = max(students, key=students.get)

average = sum(students.values()) / len(students)

def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "Fail"

print("\n--- Results ---")
for name, marks in students.items():
    print(name, ":", marks, "Grade:", grade(marks))

print("\nTopper:", topper)
print("Average:", average)
```

# Assignment-8
## Storytelling with Graphs

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [10, 20, 30, 40]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()

plt.hist(values)
plt.title("Histogram")
plt.show()
```

The histogram shows that most values are in the higher range, indicating an overall increasing trend. From the graphs, it is clear that the data is not evenly distributed, and higher categories dominate the dataset.

# Assignment-13
## Spam Classifier Thinking

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample data
messages = ["win money now", "hello friend", "free offer", "how are you"]
labels = ["spam", "not spam", "spam", "not spam"]

# Convert text to numbers
cv = CountVectorizer()
X = cv.fit_transform(messages)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test
test = cv.transform(["free money"])
print(model.predict(test))
```

### Features :

-Presence of words like “free”, “win”, “offer”

-Number of links in the message

-Length of the message

### Data Needed

-A dataset of messages (emails or SMS)

-Each message labeled as Spam or Not Spam

-Message content (text)


### Possible Mistakes

-False Positive: Normal message classified as spam

-False Negative: Spam message classified as normal

-Poor or small dataset leads to wrong predictions

# Assignment-14
## KNN in Real Life

```python
from sklearn.neighbors import KNeighborsClassifier

# Sample data (Action, Comedy)
X = [
    [5, 1],  # User A
    [4, 2],  # User B
    [1, 5],  # User C
    [2, 4]   # User D
]

# Labels (what type of movies they like)
y = ["Action", "Action", "Comedy", "Comedy"]

# Create model
model = KNeighborsClassifier(n_neighbors=3)

# Train model
model.fit(X, y)

# New user (likes Action more)
new_user = [[5, 1]]

# Prediction
prediction = model.predict(new_user)

print("Recommended Category:", prediction[0])
```

# Assignment-15
## Customer Segmentation

```python
from sklearn.cluster import KMeans

# Sample data (Income, Spending Score)
X = [
    [15, 39],
    [16, 81],
    [17, 6],
    [18, 77],
    [19, 40],
    [20, 76]
]

# Create model (3 groups)
model = KMeans(n_clusters=3)

# Train model
model.fit(X)

# Show clusters
print("Customer Groups:", model.labels_)
```

# Assignment-19
## Build a Text Cleaner

```python

text = input("Enter text: ")

# Convert to lowercase
text = text.lower()

# Remove punctuation
import string
text = text.translate(str.maketrans("", "", string.punctuation))

# Remove stopwords
stopwords = ["is", "the", "and", "in", "to", "of"]
words = text.split()

clean_words = []
for word in words:
    if word not in stopwords:
        clean_words.append(word)

# cleaned text
clean_text = " ".join(clean_words)

print("Cleaned Text:", clean_text)
```

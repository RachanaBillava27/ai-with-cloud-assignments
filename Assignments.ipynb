# Assignment-5
##  Student Data Manager

```python
students = {}

# Input for 5 students
for i in range(5):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks

topper = max(students, key=students.get)

average = sum(students.values()) / len(students)

def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "Fail"

print("\n--- Results ---")
for name, marks in students.items():
    print(name, ":", marks, "Grade:", grade(marks))

print("\nTopper:", topper)
print("Average:", average)
```

# Assignment-13
## Spam Classifier Thinking

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample data
messages = ["win money now", "hello friend", "free offer", "how are you"]
labels = ["spam", "not spam", "spam", "not spam"]

# Convert text to numbers
cv = CountVectorizer()
X = cv.fit_transform(messages)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test
test = cv.transform(["free money"])
print(model.predict(test))
```

### Features :

-Presence of words like “free”, “win”, “offer”

-Number of links in the message

-Length of the message

### Data Needed

-A dataset of messages (emails or SMS)

-Each message labeled as Spam or Not Spam

-Message content (text)


### Possible Mistakes

-False Positive: Normal message classified as spam

-False Negative: Spam message classified as normal

-Poor or small dataset leads to wrong predictions

# Assignment-19
## Build a Text Cleaner

```python

text = input("Enter text: ")

# Convert to lowercase
text = text.lower()

# Remove punctuation
import string
text = text.translate(str.maketrans("", "", string.punctuation))

# Remove stopwords
stopwords = ["is", "the", "and", "in", "to", "of"]
words = text.split()

clean_words = []
for word in words:
    if word not in stopwords:
        clean_words.append(word)

# cleaned text
clean_text = " ".join(clean_words)

print("Cleaned Text:", clean_text)
```

```python
from textblob import TextBlob

reviews = [
    "The movie was amazing",
    "I loved the acting",
    "The movie was boring",
    "Worst movie ever",
    "It was fantastic"
]

for review in reviews:
    analysis = TextBlob(review)

    if analysis.sentiment.polarity > 0:
        sentiment = "Positive"
    else:
        sentiment = "Negative"

    print(review, "->", sentiment)
```

# Assignment-27
## Image Filter Lab

```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("ani.jpg")

# Convert to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Blur
blur = cv2.GaussianBlur(gray, (5,5), 0)

# Edge Detection
edges = cv2.Canny(blur, 100, 200)

# Show Images
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img_rgb)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(gray, cmap='gray')
plt.title("Gray")

plt.subplot(2,2,3)
plt.imshow(blur, cmap='gray')
plt.title("Blur")

plt.subplot(2,2,4)
plt.imshow(edges, cmap='gray')
plt.title("Edges")

plt.show()
```
